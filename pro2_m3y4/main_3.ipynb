```python
import pandas as pd
df = pd.read_csv('./menu.csv')
```

```python
df.info()
```

```python
df['Total calories'] = df['Calories'] + df['Calories from Fat']
```

```python
filter_df = df[(df['Total calories'] < 900)]
filter_df
```

### 1 способ

```python
filter_df['Category'].value_counts()
```

```python
res = pd.merge(filter_df['Item'], filter_df['Item'], how='cross') # !
res
```

```python
res_cal = pd.merge(filter_df['Total calories'], filter_df['Total calories'], how='cross')
res_cal
```

```python
filter_df
```

```python
res_with_cal = pd.merge(res, res_cal, left_index=True, right_index=True)
res_with_cal
```

```python
res_with_cal = res_with_cal.assign(Sum=res_with_cal['Total calories_x'] + res_with_cal['Total calories_y'])
res_with_cal
```

```python
res_with_cal.drop(columns=['Total calories_x', 'Total calories_y'], inplace=True)
res_with_cal
```

```python
filtered_res_cal = res_with_cal[(res_with_cal['Sum'] < 900) & (res_with_cal['Sum'] > 600)]
filtered_res_cal
```
